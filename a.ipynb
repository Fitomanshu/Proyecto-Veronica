## 1. Descripción del Conjunto de Datos

<div class="content-box">
    <p>El modelo predictivo se construye sobre el dataset <b>PROMISE-relabeled-NICE</b>, analizando un conjunto de 622 requisitos de software estructurados.</p>
    <p>La mayoría de las variables corresponden a <b>atributos de calidad del software</b> y están codificadas de forma binaria (0/1). Esto se confirma observando que los valores mínimos y máximos son 0 y 1, representando la ausencia o presencia del atributo. Se detectó además una característica con varianza nula <i>(Other)</i>, la cual fue excluida de los análisis posteriores.</p>
</div>

<div style="display: flex; gap: 15px; justify-content: space-between;">
    <div class="metric-card" style="flex: 1;">

        <p><span class="badge-blue" style="background-color: #3498db;">49.52%</span> Funcional (1)</p>
        <p style="font-size: 0.85em; color: #7f8c8d; margin-top: 10px;"><i>El balance natural en las etiquetas permitió evitar técnicas sintéticas de oversampling.</i></p>

## 3. Diagrama de Flujo: Preprocesamiento y Validación

<div style="display: flex; flex-direction: column; align-items: center; gap: 10px; margin-top: 20px;">
    <div style="background: #fff; border-left: 5px solid #243B55; padding: 15px; width: 85%; box-shadow: 0 4px 6px rgba(0,0,0,0.05);">
        <b>1. Ingesta y Tipado:</b> Selección de 11 atributos técnicos binarios. Conversión a factor <code>category</code>. Al estar la variable objetivo naturalmente balanceada (~50/50), no fue necesario aplicar técnicas de oversampling o undersampling.
    </div>
    <div style="font-size: 1.5em; color: #243B55; font-weight: bold;">&#8595;</div>
    <div style="background: #fff; border-left: 5px solid #2980b9; padding: 15px; width: 85%; box-shadow: 0 4px 6px rgba(0,0,0,0.05);">
        <b>2. Selección de Características:</b> Extracción de predictores clave utilizando <b>V de Cramer</b> (ajuste de sesgo) e <b>Información Mutua</b> (parámetro <code>discrete_features=True</code>).
    </div>
    <div style="font-size: 1.5em; color: #2980b9; font-weight: bold;">&#8595;</div>
    <div style="background: #fff; border-left: 5px solid #8e44ad; padding: 15px; width: 85%; box-shadow: 0 4px 6px rgba(0,0,0,0.05);">
        <b>3. Partición Estratificada:</b> División en Train (70%) / Test (30%) aplicando <code>test_size=0.30</code> y <code>stratify=y</code> para preservar el balance 50/50 de la variable objetivo.
    </div>
    <div style="font-size: 1.5em; color: #8e44ad; font-weight: bold;">&#8595;</div>
    <div style="background: #fff; border-left: 5px solid #27ae60; padding: 15px; width: 85%; box-shadow: 0 4px 6px rgba(0,0,0,0.05);">
        <b>4. Esquema de Evaluación:</b> <code>StratifiedKFold</code> (<code>n_splits=10</code>, <code>shuffle=True</code>). Cálculo de estabilidad (σ) iterando 50 remuestreos (Bootstrapping) sobre datos de prueba.
    </div>
</div>

## 4. Métodos de Aprendizaje y Parámetros

<div class="content-box">
    <p>Se entrenaron dos paradigmas supervisados, preconfigurados para procesar exclusivamente variables predictoras binarias discretas (0 y 1).</p>
</div>

<div style="display: flex; gap: 20px; margin-top: 20px;">
    <div class="metric-card" style="flex: 1; border-top: 5px solid #27ae60;">
        <h4 style="color: #27ae60; margin-top: 0;">Árbol de Decisión</h4>
        <p style="font-size: 0.85em; color: #555;">Paradigma de Reglas Lógicas</p>
        <hr style="border-color: #eee;">
        <ul style="text-align: left; font-size: 0.9em; padding-left: 20px;">
            <li><b>Algoritmo:</b> <code>DecisionTreeClassifier</code></li>
            <li><b>Criterio de Partición:</b> Configurado con <code>criterion='entropy'</code> para maximizar la ganancia de información y pureza de los nodos.</li>
            <li><b>Control Estocástico:</b> <code>random_state=42</code>.</li>
            <li><b>Aplicación Extra:</b> Su vector <code>feature_importances_</code> sirvió como filtro para reducción de dimensionalidad.</li>
        </ul>
    </div>
    <div class="metric-card" style="flex: 1; border-top: 5px solid #2980b9;">
        <h4 style="color: #2980b9; margin-top: 0;">Naive Bayes</h4>
        <p style="font-size: 0.85em; color: #555;">Paradigma Probabilístico</p>
        <hr style="border-color: #eee;">
        <ul style="text-align: left; font-size: 0.9em; padding-left: 20px;">
            <li><b>Algoritmo:</b> <code>BernoulliNB</code></li>
            <li><b>Justificación:</b> Variante matemática óptima para modelar distribuciones multivariadas independientes compuestas únicamente de 0s y 1s.</li>
            <li><b>Ensambles de Filtro:</b> Ejecutado bajo 3 escenarios de entrada (Características Completas, Subconjunto del Árbol y Subconjunto de Info. Mutua).</li>
        </ul>
    </div>
</div>
