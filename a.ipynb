## 4. Métodos de Aprendizaje y Parámetros

<div class="content-box">
    <p>Se entrenaron dos paradigmas supervisados configurados para manejar un espacio híbrido de características: variables categóricas binarias y vectores de frecuencia de texto (TF-IDF).</p>
</div>
<div style="display: flex; gap: 25px; margin-top: 25px;">
    <div class="metric-card" style="flex: 1; border-top: 6px solid #27ae60;">
        <h4 style="color: #27ae60; margin-top: 0; font-size: 1.3em;">Árbol de Decisión</h4>
        <p style="font-size: 0.95em; color: #7f8c8d; font-weight: bold;">Paradigma de Reglas Lógicas</p>
        <hr style="border-color: #f1f2f6; margin: 15px 0;">
        <ul style="text-align: left; font-size: 1em; padding-left: 20px; line-height: 1.6; color: #34495e;">
            <li><b>Algoritmo:</b> <code>DecisionTreeClassifier</code></li>
            <li><b>Criterio:</b> <code>criterion='entropy'</code>. Busca maximizar la Ganancia de Información en cada división.</li>
            <li><b>Regularización:</b> <code>max_depth=15</code>. Parámetro clave para intentar mitigar el sobreajuste observado en el modelo.</li>
            <li><b>Rol Dual:</b> Actúa como clasificador y como selector de características (Feature Importance).</li>
        </ul>
    </div>
    <div class="metric-card" style="flex: 1; border-top: 6px solid #2980b9;">
        <h4 style="color: #2980b9; margin-top: 0; font-size: 1.3em;">Naive Bayes</h4>
        <p style="font-size: 0.95em; color: #7f8c8d; font-weight: bold;">Paradigma Probabilístico</p>
        <hr style="border-color: #f1f2f6; margin: 15px 0;">
        <ul style="text-align: left; font-size: 1em; padding-left: 20px; line-height: 1.6; color: #34495e;">
            <li><b>Algoritmo:</b> <code>MultinomialNB</code></li>
            <li><b>Justificación:</b> Eficiente para manejar conteos y frecuencias (TF-IDF), asumiendo independencia condicional entre términos.</li>
            <li><b>Suavizado:</b> <code>alpha=1.0</code> (Laplace Smoothing) para evitar probabilidades de cero en palabras no vistas.</li>
            <li><b>Evaluación Híbrida:</b> Ejecutado bajo 3 escenarios de filtrado para determinar la sensibilidad del modelo al ruido.</li>
        </ul>
    </div>
</div>
