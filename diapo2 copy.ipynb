<div class="content-box">
    <p>El dataset <b>fp.csv</b> contiene métricas de proyectos de software basadas en <b>Puntos Función (Function Points)</b>. 
    La variable objetivo es <b>Effort</b>, que representa el esfuerzo requerido para el desarrollo del proyecto.</p>
    <p>Durante el preprocesamiento se eliminaron las columnas no predictoras 
    <code>id</code>, <code>Project</code> y <code>YearEnd</code>. 
    Posteriormente, se realizó imputación de valores faltantes mediante la <b>mediana</b> y tratamiento de anomalías 
    en <code>TeamExp</code> y <code>ManagerExp</code>, reemplazando valores <code>-1</code> por la <b>moda</b>.</p>
</div>
<div style="display: flex; gap: 20px; justify-content: space-between;">
    <div class="metric-card" style="flex: 1; border-top: 5px solid #27ae60;">

        <p style="font-size: 1.0em;">
            <b>Imputación + Corrección de anomalías</b>
        </p>
        <p style="font-weight: bold; color: #2c3e50;">
            Valores nulos → <b>Mediana</b><br>
            Valores -1 → <b>Moda</b>
        </p>
        <p style="font-size: 0.85em; color: #7f8c8d;">
            <i>Se conservaron los registros originales evitando pérdida de información.</i>
        </p>
    </div>
    <div class="metric-card" style="flex: 1; border-top: 5px solid #f39c12;">
        <h4 style="margin-top:0; color: #141E30; font-size: 1.2em;">Escalado y Partición</h4>
        <p style="font-weight: bold;">
            Estandarización (<code>StandardScaler</code>)
        </p>
        <p>
            Escalado aplicado únicamente a variables continuas y al target.
        </p>
        <p style="margin-top: 10px;">
            <span class="badge-blue">70% Entrenamiento</span> →
            <span class="badge-blue">30% Prueba</span>

            <i>El scaler se ajustó únicamente con datos de entrenamiento para evitar data leakage.</i>
        </p>
