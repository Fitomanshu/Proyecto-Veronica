    <p>El dataset <b>fp.csv</b> contiene métricas de proyectos de software basadas en <b>Puntos Función</b>. La variable objetivo es <b>PointsAjust</b>, que representa el esfuerzo estimado (o tamaño funcional ajustado).</p>
    <p>Se eliminaron las columnas no predictoras: <code>id</code>, <code>Project</code>, <code>YearEnd</code>, <code>Language</code>. Las variables predictoras restantes son características numéricas (continuas o discretas) relacionadas con la complejidad del software.</p>
</div>

## 4. Selección de Características por Correlación de Pearson

<div class="content-box">
    <p>Se calculó la correlación absoluta de cada predictor con <b>PointsAjust</b> usando el conjunto de entrenamiento escalado. Aplicando el <b>método del codo</b> (máxima caída entre correlaciones consecutivas), se determinó el número óptimo de características a retener.</p>
</div>

    <b>Interpretación:</b> El punto de corte se sitúa en la mayor caída de la magnitud de correlación. Las características seleccionadas son aquellas con mayor relación lineal con la variable de esfuerzo.
